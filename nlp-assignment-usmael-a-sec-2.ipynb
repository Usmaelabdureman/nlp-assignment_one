# 1.2 Calculate probabilities of n-grams and find the top 10 most likely n-grams for all n.

**I calculated the probabilities as the following and print the top most common words along with their probabilities.**

```python
# Calculate probabilities of n-grams
def calculate_probabilities(ngrams):
    ngram_counts = {}
    for ngram in ngrams:
        if ngram in ngram_counts:
            ngram_counts[ngram] += 1
        else:
            ngram_counts[ngram] = 1
    total_count = len(ngrams)
    ngram_probabilities = {gram: count / total_count for gram, count in ngram_counts.items()}
    return ngram_probabilities, total_count
```

## Calculate probabilities of n-grams

```python
# Calculate probabilities for different n-grams
unigram_probabilities, unigram_total = calculate_probabilities(unigrams)
bigram_probabilities, bigram_total = calculate_probabilities(bigrams)
trigram_probabilities, trigram_total = calculate_probabilities(trigrams)
fourgram_probabilities, fourgram_total = calculate_probabilities(fourgrams)
```

```python
# Display top 10 n-grams and their probabilities for each order
def display_top_ngrams(ngram_probabilities, n, total_count):
    top_ngrams = sorted(ngram_probabilities.items(), key=lambda x: x[1], reverse=True)[:10]
    print(f"\n Top 10 {n}-grams:")
    for ngram, prob in top_ngrams:
        print(f"{ngram}: {prob}")
```

# top 10 most likely n-grams for all n and their probabilities 

```python
display_top_ngrams(unigram_probabilities, 1, unigram_total)
display_top_ngrams(bigram_probabilities, 2, bigram_total)
display_top_ngrams(trigram_probabilities, 3, trigram_total)
display_top_ngrams(fourgram_probabilities, 4, fourgram_total)
```

# 1.3 What is the probability of the sentence. "ኢትዮጵያ ታሪካዊ ሀገር ናት ". You can also try more sentences.

```python
sentence = "ኢትዮጵያ ታሪካዊ ሀገር ናት"
words = sentence.split()

# Calculate the probability of the sentence using respective n-gram probabilities
unigram_probability = 1.0
bigram_probability = 1.0
trigram_probability = 1.0
fourgram_probability = 1.0

# Calculate unigram probability of the sentences
for word in words:
    if word in unigram_probabilities:
        unigram_probability *= unigram_probabilities[word]

# Calculate bigram probability of the sentences
for i in range(len(words) - 1):
    bigram = tuple(words[i:i + 2])
    if bigram in bigram_probabilities:
        bigram_probability *= bigram_probabilities[bigram]

# Calculate trigram probability of the sentences
for i in range(len(words) - 2):
    trigram = tuple(words[i:i + 3])
    if trigram in trigram_probabilities:
        trigram_probability *= trigram_probabilities[trigram]

# # Calculate fourgram probability of the sentences
for i in range(len(words) - 3):
    fourgram = tuple(words[i:i + 4])
    if fourgram in fourgram_probabilities:
        fourgram_probability *= fourgram_probabilities[fourgram]

# Calculate the final probability using all n-gram probabilities
sentence_probability = (unigram_probability + bigram_probability + trigram_probability +fourgram_probability) / 4

print(f"Probability of the sentence: {sentence_probability}")
```

# 1.4 Generate random sentences using n-grams;
  - Here I  will generate random sentences with different length to see the impact of the increasing  ngrams.

***The aim is to create diverse sentences of varying lengths to observe the impact of different n-gram models on sentence generation***

```python
# Function to generate a random sentence using n-grams

def generate_random_sentence(ngrams, probabilities, n, max_length):
    sentence = random.choice(list(ngrams)).split()
    while len(sentence) < max_length:
        prev_words = ' '.join(sentence[-1])
        choices = [(key, value) for key, value in probabilities.items()]
        if not choices:
            break
        next_word = np.random.choice([word for word, _ in choices], p=[prob for _, prob in choices])
        sentence.append(next_word)
    return ' '.join(sentence)
```

# 2. Evaluate these Language Models Using Intrinsic Evaluation Method (4%)
I will use ***PERPLEXITY*** as intrinsic evaluation method

```python
def calculate_perplexity(probabilities, test_set):
    N = len(test_set)
#     total_probability = 1.0
    log_total_probability = 0.0

    for i in range(N - 1):  # Loop through each pair of consecutive words
        # Calculate the conditional probability of the next word given the current word
        current_word = test_set[i]
        if current_word in probabilities:
            conditional_prob = probabilities[current_word]
#             total_probability *= 1/conditional_prob
            log_total_probability += math.log2(conditional_prob)
        else:
            # Handle unseen word sequences with a small probability
#             total_probability *= (1/0.0001)  # Set a small probability for unseen sequences
            log_total_probability += math.log2(0.0001)
#     perplexity = pow(total_probability, 1 / N)
    average_log_probability = log_total_probability / (N - 1)
    perplexity = pow(2, -average_log_probability)
    return perplexity
```
